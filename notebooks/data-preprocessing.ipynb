### Outlier Detection

```python
from scipy import stats
import numpy as np

z_score = stats.zscore(df.select_dtypes(include=[float]))
abs_z_scores = np.abs(z_score)
outliers = (abs_z_scores > 3).all(axis=1)
df_rm_outlier = df[~outliers]
df_rm_outlier.head()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df_rm_outlier)
plt.xticks(rotation=90)
plt.show()
```

### Feature Scaling & Balancing the Dataset

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_feats = scaler.fit_transform(df.drop(columns=['Diabetes_012']))
df_scaled = pd.DataFrame(scaled_feats, columns=df.columns.drop('Diabetes_012'))
df_scaled['Diabetes_012'] = df['Diabetes_012'].values
df_scaled.head()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X, y = df_scaled.drop(columns=['Diabetes_012']), df_scaled['Diabetes_012']
X_resampled, y_resampled = smote.fit_resample(X, y)
df_resampled = pd.DataFrame(X_resampled, columns=df_scaled.columns.drop('Diabetes_012'))
df_resampled['Diabetes_012'] = y_resampled
```

```python
df_resampled.hist(figsize=(15, 10), bins=20, layout=(5, 5))
plt.tight_layout()
plt.show()
```

```python
fldr = os.environ['processed_fldr']
file = 'preprocessed_data.csv'
df_resampled.to_csv(os.path.join(fldr, file), index=False)
```
